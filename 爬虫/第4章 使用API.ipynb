API也叫做网络应用
其返回的值值得格式往往为：API 会用XML（eXtensible Markup Language， 可扩展标记语言） 或JSON（JavaScript Object Notation，JavaScript 对象表示）格式返回服务器响应的信息

## 4.1 API概述
API 之所以叫API 而不是叫网站的原因，其实是首先API 请求使用非常严谨的语法，其次API 用JSON 或XML 格式表示数据，而不是HTML 格式。

## 4.2 API通用规则
GET 就是你在浏览器中输入网址浏览网站所做的事情。
POST 基本就是当你填写表单或提交信息到网络服务器的后端程序时所做的事情。
PUT 请求用来更新一个对象或信息。
DELETE 用于删除一个对象。

```python
import json
from urllib.request import urlopen

def getcountry(idaddress):
    response = urlopen("http://freegeoip.net/json/"+idaddress).read().decode('utf-8')
    responsejson = json.loads(response)
    return responsejson.get("country_code")
print(getcountry("50.78.253.58"))
```
