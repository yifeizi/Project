## pandas 删除缺失值
使用dropna（）方法
DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
参数说明：

axis：默认为 0，表示逢空值剔除整行，如果设置参数 axis＝1 表示逢空值去掉整列。
how：默认为 'any' 如果一行（或一列）里任何一个数据有出现 NA 就去掉整行，如果设置 how='all' 一行（或列）都是 NA 才去掉这整行。
thresh：设置需要多少非空值的数据才可以保留下来的。
subset：设置想要检查的列。如果是多个列，可以使用列名的 list 作为参数。
inplace：如果设置 True，将计算得到的值直接覆盖之前的值并返回 None，修改的是源数据。

## pandas清洗格式错误数据
to_datetime()将错误日期数据转化为标准日期格式

```python
data = {
    'Data':['2020/12/01','20201202','20201203'],
    'nums':[50,55,45]
}
date = pd.DataFrame(data,index=['day1','day2','day3'])
date
```

```python
date['Data'] = pd.to_datetime(date['Data'])
date
```

## pandas清洗重复数据
使用duplicated（）和drop_duplicates（）


```python
person = {
    'name':['Google','Runoob','Runoob','Taobao'],
    "age":[50,40,40,23]
}
df5 = pd.DataFrame(person,index=['one','two','three','four'])
```

```python
df5
```

```python
df5.duplicated()
```

```python
df6 = df5.drop_duplicates()
df6
```
