```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.tableau.com/zh-cn/solutions/customer/hellofresh-boosts-digital-marketing-campaigns-and-increases-conversion-rates"
html = requests.get(url)

#print(html.text)
bsobj = BeautifulSoup(html.text)

for i in bsobj.findAll("p"):
    print(i.get_text())
    with open("bi1.txt",mode="a",encoding='utf-8') as f:
        f.write(i.get_text())
```

```python
url_l = "https://www.tableau.com/zh-cn/solutions/customer/lenovo-increases-reporting-efficiency-across-enterprise"

html_l = requests.get(url_l)

#print(html_l.text)

bsobj_l = BeautifulSoup(html_l.text)

for j in bsobj_l.findAll("p"):
    print(j.get_text())
    with open("lenovo.txt",mode="a",encoding="utf-8") as f:
        f.write(j.get_text())
```
