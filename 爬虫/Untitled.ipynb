```python
def countdown(n):
    print("start!")
    while n > 0:
        yield n
        n -= 1
    print("done!")

c = countdown(3)
c
```

```python
def foo():
    print("starting......")
    while True:
        res = yield 2#遇到yield函数暂停运行等待next（）方法的调用，函数继续运行
        print("res:",res)#yield是返回值而不能进行赋值
        
g = foo()
print("frist carry out......")
print(next(g))
print("*"*100)

print("second carry out......")
print(next(g))
print("^"*100)
```
