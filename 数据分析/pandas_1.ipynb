## 分组
“分组方式”指的是涉及以下一个或多个步骤的流程：
    ·通过一些标准来划分数据
    ·向每一个组单独应用一些函数
    ·将这些结果合并起来

```python
df = pd.DataFrame(
     {
         "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
         "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
         "C": np.random.randn(8),
         "D": np.random.randn(8),
     }
)
df
```

```python
#分组，然后将sum（）函数应用于结果组。
df.groupby("A").sum()
```

```python
#按多列分组形成一个层次索引，同样，我们可以应用sum（）函数。
df.groupby(['A','B']).sum()
```

## 重塑


```python
tuples = list(
        zip(
             *[
                 ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
                 ["one", "two", "one", "two", "one", "two", "one", "two"],
            ]
         )
 )
tuples
```

```python
index = pd.MultiIndex.from_tuples(tuples,names=['first','second'])
df = pd.DataFrame(np.random.randn(8,4),index=index,columns=['A','B','C','D'])
df
```

```python
#stack（）方法“压缩”数据帧列中的一个级别。
#对于“堆叠”数据帧或序列（具有多索引作为索引），stack（）的逆操作为unstack（），默认情况下，该操作将取消最后一级的堆栈：
stacked = df.stack()
stacked
```

```python
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

## 时间序列
pandas具有简单、强大且高效的功能，用于在频率转换期间执行重采样操作（例如，将第二个数据转换为5分钟数据）。这在金融应用中非常常见，但不限于此。

```python
rng = pd.date_range("20210101",periods=100,freq='s')
rng
```

```python
ts = pd.Series(np.random.randint(0,500,len(rng)),index=rng)
ts
```

```python
ts.resample("1Min").sum()
```

```python
ts_utc = ts.tz_localize("UTC")
ts_utc
```

```python
ts_utc.tz_convert("US/Eastern")
```
